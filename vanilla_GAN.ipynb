```python
for _ in range(100):
    
    X_true = create_real_data(size=200)
    y_true = np.ones((len(X_true), 1))  
    generator_inputs = latent_space_points(5, len(X_true))
    X_fake = model_generator.predict(generator_inputs)
    y_fake = np.zeros((len(X_true), 1))
    X_dis = np.concatenate((X_fake, X_true))
    y_dis = np.concatenate((y_fake, y_true))
    
    model_discriminator.trainable=True
    model_discriminator.fit(X_dis, y_dis, verbose=0, epochs=100)
    model_discriminator.trainable=False
    
    X_gan = latent_space_points(5, 200)
    y_gan = np.ones((200, 1))
    gan_model.fit(X_gan, y_gan, verbose=0, epochs=100)
    
    visualise_generator_points(model_generator, 5, 200)
```
